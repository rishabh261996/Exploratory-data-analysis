# Insights about teams

```python
# Number of matches won by teams
sns.countplot(x='winner',data=data1, order = data1['winner'].value_counts().index[:5]) # top 5 teams in 
plt.xticks(rotation =90)
plt.xlabel("Winner team",fontsize=14)
plt.ylabel("Number of winss",fontsize=14)
plt.title("Showing  top 5 team's performance ",fontsize =18)
```


Teams  like Mumbai and Chennai Super kings have won maximum number of matches in the ipl.


```python
# Dependency on toss decision
sns.countplot(x='toss_winner',data=data1, order = data1['winner'].value_counts().index[:5],hue='toss_decision')
plt.xticks(rotation=90)
plt.xlabel("Number of matches won by the home team",fontsize=12)
plt.ylabel("teams",fontsize=12)
plt.title("Showing team's performance with respect to toss decision",fontsize =18)
```

Teams winner the toss are more likely to choose field first. However, Teams like Royal Challenger Banglore are going for field more often than others. Whereas, Chennai Super kings is an expection with its preference is more for fielding than batting first.


```python
data1[data1.toss_winner ==data1.winner]['winner'] .count() /data1['winner'].count()
```


52 % of the times a team winning toss also won the match. So , toss could be a factor deciding the result of the match. As team winning tosses are likely to make decision that give undue advantage over other

```python
# home team winnning the match
data_homewinner = data1[data1['team1'] == data1['winner']]
data_homewinner['winner'].value_counts().sort_values(ascending=False)[:5].plot(kind='pie',autopct='%.2f')
plt.title("Showing team's performance in their grounds",fontsize =18)
```

```python
# host team winnning the match
data_hostwinner = data1[data1['team2'] == data1['winner']]
data_hostwinner['winner'].value_counts()[:5].plot(kind='pie',autopct='%.2f')
plt.title("Showing team's performance in other grounds",fontsize =18)
```

###  Comparing both home and hostwinner:
Teams like Mumbai indians and Chennai Super Kings performs well on both home and other grounds. While KKR which wins highest number of matches in home conditons fails to replicate the same in foreign conditions.

```python
#team score per wicket
Df1  = pd.DataFrame(data2.groupby(['batting_team'])['total_runs'].sum())
Df1.reset_index(inplace=True)
Df1.rename(columns= {'batting_team':'team'},inplace=True)
Df2  = pd.DataFrame(data2.groupby(['bowling_team'])['player_dismissed'].count().sort_values(ascending =False))
Df2.reset_index(inplace=True)
Df2.rename(columns= {'bowling_team':'team'},inplace=True)
Df = pd.merge(Df1,Df2,how='inner')
```

```python
# a variable taking into account the performance of the team in both wickets and runs ( feature engineering)
Df['runs_scored_wicket']= (Df['total_runs']*Df['player_dismissed'])**0.5
Df[['team','runs_scored_wicket']].sort_values(by='runs_scored_wicket',ascending =False)[:5].plot(kind='barh',x='team',y='runs_scored_wicket',color=(0.5,0.5,0.1))
plt.title("Performance from wickets taken and runs scored")
```

The above graph illustrated the performance of top 5 teame in terms of wickets and runs it score. On that basis, Mumbai Indian is the top performer in this too. However, the rest 4 teams are not much behind it.

# Insights about players

```python
#top 5 match winners
sns.countplot(x= 'player_of_match',data=data1, order = data1['player_of_match'].value_counts().index[:5])
plt.xlabel("Players",fontsize=14)
plt.ylabel("Number of match winning awards",fontsize=14)
plt.title("Top 5 Match Winners",fontsize =18)
```


Match winners are players deciding the result of the match. So, a company would be willing to sponspor players having role in match result. Here, players such as CH Gayle MS Dhoni and RG Sharma are in the top 5 players in terms of number of match winner awards.

```python
# top run getters:
data2.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)[:5].plot(kind='bar',color =(0,0.8,0.5))
plt.xlabel("Players",fontsize=14)
plt.ylabel("Run scored",fontsize=14)
plt.title("Top 5 Batsmen",fontsize =18)
```

```python
#top wicket takers:
data2.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False)[:5].plot(kind='bar',color =(1,0.5,0.1))
plt.xlabel("Bowlers",fontsize=14)
plt.ylabel("Wickets taken",fontsize=14)
plt.title("Top 5 Wicket Takers",fontsize =18)
```


Top Players in terms of aggregate runs are V kohli, Sk raina, RG Sharma, DA Warner, and S Dhawan.

Top Players in terms of wickets are SL Malinga, DJ Bravo, A Mishra , and P Chawla.

```python
# top catchers:
data2['fielder'].value_counts()[:5].plot(kind='barh',color = (0.1,0.5,1))
plt.xlabel("Catching fielders",fontsize=14)
plt.ylabel("Number of catches caught",fontsize=14)
plt.title("Top 5 Catchers",fontsize =18)
```


Top players in terms of catching are AB de villiers, SK Raina, RK Uttappa, KD Karthik and MS Dhoni.

```python
# Top-performing allrounders:
df1 = pd.DataFrame(data2.groupby('bowler')['player_dismissed'].count())
df1.reset_index(inplace=True)
df1.rename(columns= {'bowler':'player','player_dismissed':'wickets'},inplace=True)
df2 = pd.DataFrame(data2.groupby('batsman')['batsman_runs'].sum())
df2.reset_index(inplace=True)
df2.rename(columns ={'batsman':'player','batsman_runs':'runs'},inplace=True)
df = pd.merge(df1,df2,how='inner')
#defining a columns which counts efforts in bowling and batting equally
df['runs_scored_withwicket'] = (df['wickets']*df['runs'])**0.5 
df[['player','runs_scored_withwicket']].sort_values(by= 'runs_scored_withwicket',ascending=False)[:5].plot( x= 'player',
                                                                                                           y='runs_scored_withwicket',
                                                                                                           kind='barh',color=(0.5,0.5,0.6))
plt.title("Showing performance of each player from both batting and bowling")
```


Players like Shane Watson , DJ Bravo and RA Jadeja are best performing all rounders
